```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('study_performance.csv')

jumlah_siswa_per_etnis = df['race_ethnicity'].value_counts()

jumlah_siswa_per_etnis = jumlah_siswa_per_etnis.sort_index()

plt.bar(jumlah_siswa_per_etnis.index, jumlah_siswa_per_etnis.values,color ='skyblue')

plt.title('Jumlah Siswa Pada Tiap Kelompok Etnis')
plt.xlabel('Kelompok Etnis')
plt.ylabel('Jumlah Siswa')

plt.show()
```
